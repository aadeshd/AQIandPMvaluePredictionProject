```python
#To calculate total PM 2.5 in a grid
PM_25=PMdata[['PM25_IND','PM25_OTH','PM25_POW','PM25_RES','PM25_TRN','PM25_WB']].sum(axis=1)
#concatinate total pm 2.5 datafarme with other columns
PMdata = pd.concat([PMdata, PM_25], axis=1, sort=False)
#rename total pm 2.5 column as PM
PMdata.rename(columns={0:"PM"},inplace=True)
PMdata['Trees']=0
PMdata.head()
```

```python
PMdata['LATL']=0
PMdata['LATR']=0
PMdata['LONGL']=0
PMdata['LONGR']=0
for i in range(len(PMdata)):
    #By observation we can see each grid is o.oo5 unit apart from other.
    #Thus we consider a given point as the center of the grid.
    w=PMdata.loc[i,'LAT']-0.0025 #left 0.0025 from center
    PMdata.loc[i,"LATL"]=w
    x=PMdata.loc[i,'LAT']+0.0025 #right 0.0025 from center
    PMdata.loc[i,"LATR"]=x
    y=PMdata.loc[i,"LONG"]-0.0025 #up 0.0025 from center
    PMdata.loc[i,"LONGL"]=y
    z=PMdata.loc[i,"LONG"]+0.0025 #below 0.0025 from center
    PMdata.loc[i,"LONGR"]=z
```

```python
PMdata.drop('LONG',axis=1,inplace=True)
```

```python
PMdata.drop('LAT',axis=1,inplace=True)
```

```python
start = time.process_time()
latid=0
for i in Treepart1.index:
    count=0
    a=Treepart1.loc[i,'latitude'] #latitude of a tree
    b=Treepart1.loc[i,'longitude']#longitude of a tree
    for j in range(latid,len(PMdata)):
        if(PMdata.loc[j,'LATL']<=a<=PMdata.loc[j,'LATR']):
            if(PMdata.loc[j,'LONGL']<=b<=PMdata.loc[j,'LONGR']): #it will check if
                PMdata.loc[j,"Trees"]=PMdata.loc[j,"Trees"]+1 #if above condition satisfies it append tree column of that row as +1
                latid=j
                break #it will break the loop if condition satisfies
print(time.process_time() - start)
```
